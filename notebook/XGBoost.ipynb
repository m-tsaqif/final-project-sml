# Splitting Dataset

```python
X = df.drop(columns=["label"])
y = df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Feature Engineering & Data Preprocessing

```python
categoric_pipeline = Pipeline([
    ('encoder',utils.TargetFeatureEncoder(to_encode=categoric_features))
])

numeric_pipeline = Pipeline([
    ('yeojohnson', utils.YeoJohnsonTransformer(columns=numeric_features)),
    ('scaler', utils.StandardScalerTransformer(columns=numeric_features))
])

preprocessor = ColumnTransformer([
    ('categoric', categoric_pipeline, categoric_features),
    ('numeric', numeric_pipeline, numeric_features)
], verbose_feature_names_out=False)
```

```python
# Encode label menjadi angka
label_encoder = LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
y_test_enc = label_encoder.transform(y_test)
```

```python
# Compute sample weights untuk menangani ketidakseimbangan kelas
sample_weights = compute_sample_weight('balanced', y_train_enc)
```

# Feature Selection

```python
# Mutual Information
preprocessor.fit(X_train, y_train_enc)
mi_results = utils.plot_mutual_info_scores(preprocessor, X_train, y_train_enc)
```

# Model Training

```python
# Training
NUM_CLASSES = len(label_encoder.classes_)
TOP_K = 26

xgb_model = XGBClassifier(
    objective='multi:softprob',
    num_class=NUM_CLASSES,
    eval_metric='mlogloss',
    random_state=42
)

feature_selector = SelectKBest(score_func=mutual_info_classif, k=TOP_K)

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('feature_selector', feature_selector),
    ('algo', utils.StratifiedKFoldMulticlass(
        estimator=xgb_model,
        n_folds=3,
        method='predict_proba',
        random_state=42
    ))
])

pipeline.fit(
    X_train, 
    y_train_enc,
    algo__sample_weight=sample_weights
)
```

```python
# Prediction
y_pred_enc = pipeline.predict(X_test)
y_proba = pipeline.predict_proba(X_test)
```

```python
# Evaluation
evaluator = utils.ClassificationEvaluator(
    y_true=y_test_enc,
    y_pred=y_pred_enc,
    y_proba=y_proba,
    label_encoder=label_encoder
)

evaluator.plot_confusion_matrix(title="XGBoost - Confusion Matrix")
evaluator.plot_roc_curves(title="XGBoost - ROC-AUC Curves")
evaluator.plot_probability_distribution(title="XGBoost - Predicted Probabilities Distribution by True Class")
```

# Hyperparameter Tuning

```python
# Konstanta
NUM_CLASSES = len(label_encoder.classes_)
TOP_K = 26
N_TRIALS = 10

# Feature selector
feature_selector = SelectKBest(score_func=mutual_info_classif, k=TOP_K)

# Fungsi objective dimodifikasi supaya bisa dipanggil manual
def manual_objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 15),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'gamma': trial.suggest_float('gamma', 0, 5),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 5),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 5),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'objective': 'multi:softprob',
        'num_class': NUM_CLASSES,
        'eval_metric': 'mlogloss',
        'use_label_encoder': False,
        'random_state': 42
    }

    xgb_model = XGBClassifier(**params)
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('feature_selector', feature_selector),
        ('clf', xgb_model)
    ])

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    losses = []

    for train_idx, valid_idx in skf.split(X_train, y_train_enc):
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[valid_idx]
        y_tr, y_val = y_train_enc[train_idx], y_train_enc[valid_idx]
        sw_tr = sample_weights[train_idx]

        pipeline.fit(X_tr, y_tr, clf__sample_weight=sw_tr)
        y_pred_proba = pipeline.predict_proba(X_val)
        losses.append(log_loss(y_val, y_pred_proba))

    return np.mean(losses)

# Manual progress bar loop
study = optuna.create_study(direction='minimize')
for _ in tqdm(range(N_TRIALS), desc="Optuna Trials"):
    study.optimize(manual_objective, n_trials=1, catch=(Exception,))
```

```python
print("Best trial:")
print("  Value (log loss):", study.best_value)
print("  Params:")
for key, value in study.best_trial.params.items():
    print(f"    {key}: {value}")
```

# Evaluation

```python
# Training
NUM_CLASSES = len(label_encoder.classes_)
TOP_K = 26

best_params = {
    'n_estimators': study.best_trial.params['n_estimators'],
    'max_depth': study.best_trial.params['max_depth'],
    'learning_rate': study.best_trial.params['learning_rate'],
    'subsample': study.best_trial.params['subsample'],
    'colsample_bytree': study.best_trial.params['colsample_bytree'],
    'gamma': study.best_trial.params['gamma'],
    'reg_alpha': study.best_trial.params['reg_alpha'],
    'reg_lambda': study.best_trial.params['reg_lambda'],
    'min_child_weight': study.best_trial.params['min_child_weight'],
    'objective': 'multi:softprob',
    'num_class': NUM_CLASSES,
    'eval_metric': 'mlogloss',
    'use_label_encoder': False,
    'random_state': 42
}

xgb_model = XGBClassifier(**best_params)

feature_selector = SelectKBest(score_func=mutual_info_classif, k=TOP_K)

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('feature_selector', feature_selector),
    ('algo', utils.StratifiedKFoldMulticlass(
        estimator=xgb_model,
        n_folds=3,
        method='predict_proba',
        random_state=42
    ))
])

pipeline.fit(
    X_train, 
    y_train_enc,
    algo__sample_weight=sample_weights
)
```

```python
# Prediction
y_pred_enc = pipeline.predict(X_test)
y_proba = pipeline.predict_proba(X_test)
```

```python
# Evaluation
evaluator = utils.ClassificationEvaluator(
    y_true=y_test_enc,
    y_pred=y_pred_enc,
    y_proba=y_proba,
    label_encoder=label_encoder
)

evaluator.plot_confusion_matrix(title="XGBoost - Confusion Matrix")
evaluator.plot_roc_curves(title="XGBoost - ROC-AUC Curves")
evaluator.plot_probability_distribution(title="XGBoost - Predicted Probabilities Distribution by True Class")
```

# SHAP

```python
# --- 1. Preprocessing dan Feature Selection ---
# Transformasi data dengan y (untuk TargetEncoder)
X_train_transformed = preprocessor.fit_transform(X_train, y_train_enc)
X_test_transformed = preprocessor.transform(X_test)

# Pastikan hasil transformasi berupa array
X_train_transformed = pd.DataFrame(X_train_transformed)
X_test_transformed = pd.DataFrame(X_test_transformed)

# Dapatkan nama fitur setelah preprocessing
pre_feat_names = preprocessor.get_feature_names_out()
X_train_transformed.columns = pre_feat_names
X_test_transformed.columns = pre_feat_names

# Seleksi fitur
feature_selector = SelectKBest(score_func=mutual_info_classif, k=TOP_K)
X_train_selected = feature_selector.fit_transform(X_train_transformed, y_train_enc)
X_test_selected = feature_selector.transform(X_test_transformed)

# Ambil nama fitur yang dipilih
selected_mask = feature_selector.get_support()
selected_feat_names = pre_feat_names[selected_mask]
```

```python
# --- 2. Fit Model secara Langsung ---
xgb_model_direct = XGBClassifier(
    objective='multi:softprob',
    num_class=len(np.unique(y_train_enc)),
    eval_metric='mlogloss',
    random_state=42
)

xgb_model_direct.fit(X_train_selected, y_train_enc, sample_weight=sample_weights)
```

```python
# Gunakan object Explainer (bukan .shap_values)
explainer = shap.Explainer(xgb_model_direct, X_train_selected)
shap_values = explainer(X_test_selected)  # Ini object 'Explanation'

class_names = label_encoder.classes_

# Plot semua kelas (SHAP values multi-class)
shap.summary_plot(shap_values, X_test_selected, feature_names=selected_feat_names, class_names=class_names)

# Plot hanya untuk kelas tertentu (misalnya kelas ke-0 = label asli?)
shap.summary_plot(shap_values[:, :, 0], X_test_selected, feature_names=selected_feat_names)
```
